```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data_model2 = pd.read_csv('avgword2vec_cbow.csv')
```

```python
train_data_model2.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'train_id'], inplace=True)
train_data_model2
```

```python
data = train_data_model2.copy()
from sklearn.model_selection import train_test_split

y = data['price']
X = data.drop(columns=['price'])


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
X_train = X_train[:10000]
y_train = y_train[:10000]
X_test = X_test[:3500]
y_test = y_test[:3500]
```

```python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components=100)
X_K_train = kpca.fit_transform(X_train)
X_K_test = kpca.transform(X_test)
```

```python
y1 = model1.predict(X_test)
y2 = model2.predict(X_K_test)
y3 = model3.predict(X_K_test)
y4 = model4.predict(X_test)
y5 = model5.predict(X_K_test)
model6.fit(X_K_train, y_train)
y6 = model6.predict(X_K_test)
```
